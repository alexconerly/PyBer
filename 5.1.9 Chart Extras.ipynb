### Add Error Bars to a Line Chart

```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Get the standard deviation of the values in the y-axis.
SD = statistics.stdev(y_axis)
SD
```

```python
#  USE:  plt.errorbar(x, y, yerr=<value>) to add errorbars

plt.errorbar(x_axis, y_axis, yerr=SD, capsize=5)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=SD, capsize=3)
plt.show()
```

### Add Error Bars to a Bar Chart

```python
#MATLAB approach to creating errorbars.  
#Also, note you don't have to write .errorbar like you did for linecharts
plt.bar(x_axis, y_axis, yerr=SD, capsize=3)
```

```python
#Object oriented approach to creating errorbars
import numpy as np
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=SD, capsize=3)
plt.show()
```

### Change the Major Ticks

```python
import numpy as np
plt.barh(x_axis, y_axis)

plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
```

```python
import numpy as np
np.arange(0, 51, step=5.0)
```

```python
#Get same graph using object-oriented approach

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()


#note, if we had numbers for the y axis we could also set yticks
```

### Add Minor Ticks

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
